1.Extract Headers from excel

```python
import os
import pandas as pd

folder_path = r"D:\Generation Course\CW1\CSV data"  # Update with your folder path
output_file = r"D:\Generation Course\CW1\headers_list.txt"  # Path for the output file

with open(output_file, 'w') as outfile:
    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            df = pd.read_csv(file_path)

            # Write headers to the output file
            outfile.write(f"File: {file}\n")
            outfile.write(f"Headers: {', '.join(df.columns)}\n")
            outfile.write("-" * 40 + "\n")
print(f"Headers extracted and saved to {output_file}")



```

Slide 7 heat map sex and job title

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# File paths
path = "D:/Generation Course/CW1/CSV data/"

# Load the datasets
hr_employee = pd.read_csv(path + "HR Employee.csv")
sick_leave_location = pd.read_csv(path + "Sick Leave by Location.csv")
sick_leave_department = pd.read_csv(path + "Sick leave Department and job titles.csv")

# Preview the datasets
print("HR Employee Data:")
print(hr_employee.head())

print("\nSick Leave by Location Data:")
print(sick_leave_location.head())

print("\nSick Leave Department and Job Titles Data:")
print(sick_leave_department.head())

# Standardize column names
hr_employee.rename(columns={"Job Title": "JobTitle"}, inplace=True)
sick_leave_location.rename(columns={"Job Title": "JobTitle"}, inplace=True)

# Strip whitespace from JobTitle
hr_employee['JobTitle'] = hr_employee['JobTitle'].str.strip()
sick_leave_location['JobTitle'] = sick_leave_location['JobTitle'].str.strip()
sick_leave_department['JobTitle'] = sick_leave_department['JobTitle'].str.strip()

# Verify the changes
print("\nColumn Names in HR Employee:")
print(hr_employee.columns)

print("\nColumn Names in Sick Leave Location:")
print(sick_leave_location.columns)

print("\nColumn Names in Sick Leave Department:")
print(sick_leave_department.columns)

hr_location_merge = pd.merge(
    hr_employee,
    sick_leave_location,
    on="JobTitle",
    how="inner"
)

print("\nPreview of HR + Location Merge:")
print(hr_location_merge.head())

merged_data = pd.merge(
    hr_location_merge,
    sick_leave_department,
    on="JobTitle",
    how="inner"
)

print("\nPreview of Fully Merged Data:")
print(merged_data.head())

# Check for missing values
print("\nMissing Values in Merged Data:")
print(merged_data.isnull().sum())

# Drop rows with missing values in critical columns
merged_data.dropna(subset=["Gender", "MaritalStatus", "Department", "SickLeaveHours_y"], inplace=True)

# Verify cleaned data
print("\nCleaned Merged Data:")
print(merged_data.head())

# Apply logarithm transformation to sick leave hours
merged_data['Log_SickLeaveHours'] = np.log1p(merged_data['SickLeaveHours_y'])

# Verify the transformation
print("\nPreview of Transformed Data:")
print(merged_data[['JobTitle', 'Department', 'SickLeaveHours_y', 'Log_SickLeaveHours']].head())

# Print column names to verify the correct column name for sick leave hours
print("\nColumn Names in Merged Data:")
print(merged_data.columns)

# Update the pivot table to use the correct column name
heatmap_data = merged_data.pivot_table(
    values="SickLeaveHours_y",  # Use the correct column name
    index="Department",
    columns=["Gender", "MaritalStatus"],
    aggfunc="mean"
)

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".1f",
    cmap="coolwarm"
)
plt.title("Sick Leave Hours by Department, Gender, and Marital Status", fontsize=14)
plt.xlabel("Gender and Marital Status", fontsize=12)
plt.ylabel("Department", fontsize=12)
plt.tight_layout()
plt.show()



```

Following are SQL data codes - Gender,Marital Status & Shift Analysis for Prod Tech

```python
SELECT 
    e.Gender, 
    e.MaritalStatus, 
    s.Name AS ShiftName, 
    SUM(e.SickLeaveHours) AS TotalSickLeaveHours
FROM HumanResources.Employee e
JOIN HumanResources.EmployeeDepartmentHistory edh ON e.BusinessEntityID = edh.BusinessEntityID
JOIN HumanResources.Shift s ON edh.ShiftID = s.ShiftID
WHERE e.JobTitle LIKE '%Production Technician%'
GROUP BY e.Gender, e.MaritalStatus, s.Name
ORDER BY TotalSickLeaveHours DESC;
```

HR Department grouping of roles

```python
SELECT 
    d.Name AS DepartmentName, 
    e.JobTitle, 
    SUM(e.SickLeaveHours) AS TotalSickLeaveHours, 
    COUNT(e.BusinessEntityID) AS NumberOfEmployees, 
    AVG(e.SickLeaveHours) AS AverageSickLeaveHours
FROM HumanResources.Employee e
JOIN HumanResources.EmployeeDepartmentHistory edh ON e.BusinessEntityID = edh.BusinessEntityID
JOIN HumanResources.Department d ON edh.DepartmentID = d.DepartmentID
WHERE e.JobTitle LIKE '%Production Technician%' AND edh.EndDate IS NULL
GROUP BY d.Name, e.JobTitle
ORDER BY TotalSickLeaveHours DESC;
```

Sick leave analysis by department and average hours

```python
SELECT 
    d.Name AS DepartmentName, 
    e.JobTitle, 
    SUM(e.SickLeaveHours) AS TotalSickLeaveHours, 
    COUNT(e.BusinessEntityID) AS NumberOfEmployees, 
    AVG(e.SickLeaveHours) AS AverageSickLeaveHours
FROM HumanResources.Employee e
JOIN HumanResources.EmployeeDepartmentHistory edh ON e.BusinessEntityID = edh.BusinessEntityID
JOIN HumanResources.Department d ON edh.DepartmentID = d.DepartmentID
WHERE e.JobTitle LIKE '%Production Technician%' AND edh.EndDate IS NULL
GROUP BY d.Name, e.JobTitle
ORDER BY TotalSickLeaveHours DESC;
```

Slide 16 Sick leave & Scrap Qty by Dept &  Work Order Qty

```python
SELECT 
    e.BusinessEntityID,
    e.JobTitle,
    e.SickLeaveHours,
    d.Name AS DepartmentName,
    wor.WorkOrderID,
    wo.ProductID,
    wo.ScrappedQty,
    wor.ActualResourceHrs,
    wor.ScheduledStartDate,
    wor.ActualCost
FROM HumanResources.Employee e
JOIN HumanResources.EmployeeDepartmentHistory edh ON e.BusinessEntityID = edh.BusinessEntityID
JOIN HumanResources.Department d ON edh.DepartmentID = d.DepartmentID
JOIN Production.WorkOrderRouting wor ON d.DepartmentID = wor.LocationID
JOIN Production.WorkOrder wo ON wor.WorkOrderID = wo.WorkOrderID
WHERE e.SickLeaveHours > 0

ORDER BY wo.ScrappedQty DESC, wor.ScheduledStartDate ASC;
```

Slide 16 Sick Leave Correlation with Work Order Reource Utlisastion and Timelines

```python
SELECT 
    e.BusinessEntityID,
    e.JobTitle,
    e.SickLeaveHours,
    wor.WorkOrderID,
    wor.ResourceID,
    wor.ActualResourceHrs,
    wor.ScheduledHours,
    wor.StartDate,
    wor.EndDate
FROM HumanResources.Employee e
JOIN Production.WorkOrderRouting wor ON e.BusinessEntityID = wor.ResourceID
WHERE e.SickLeaveHours > 0
  AND wor.StartDate BETWEEN '2025-01-01' AND '2025-12-31'
ORDER BY wor.StartDate;
```
